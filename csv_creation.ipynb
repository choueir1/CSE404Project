```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
apple= yf.download("AAPL", start = "2020-01-02", end = "2024-12-31")
google = yf.download("GOOG", start = "2020-01-02", end = "2024-12-31")
apple.head()
```

```python
yf.__version__
```

```python
apple = pd.DataFrame(apple)
google = pd.DataFrame(google)
```

```python
df = pd.concat([apple, google], axis=0)
```

```python
google
```

```python
df.columns
```

```python
tickers = ["MSFT", "AMZN", "TSLA", "NVDA", "META", "JPM", "V", "BABA", "TSM",
    "SPY", "QQQ", "ARKK", "VOO", "DIA", "XLF",
    "BTC-USD", "ETH-USD", "DOGE-USD", "ADA-USD", "XRP-USD", "SOL-USD",
    "^GSPC", "^DJI"]

print(tickers)
```

```python
for ticker in tickers:
    stock_df = yf.download(ticker, start = "2020-01-02", end = "2024-12-31")
    df = pd.concat([df, stock_df], axis=0)
```

```python
multi_index = pd.MultiIndex.from_tuples(df.columns)
```

```python
flat_columns = ['_'.join(col) for col in multi_index]
```

```python
df.columns = flat_columns
```
